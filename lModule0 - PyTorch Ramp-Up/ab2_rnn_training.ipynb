```python
import torch
import torch.nn as nn
import torch.optim as optim

torch.manual_seed(0)
```

1. Define RNN model

```python
class MyRNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.rnn = nn.RNN(input_size=10, hidden_size=20, batch_first=True)
    self.fc = nn.Linear(20,1)

  def forward(self,x):
    out, _ = self.rnn(x)
    out = out[:,-1,:]
    return self.fc(out)

model = MyRNN()
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()
```

2. Dummy training loop

```python
for epoch in range(21):
    x = torch.rand(32,1,10)
    y = torch.rand(32,1)

    optimizer.zero_grad()
    preds = model(x)
    loss = criterion(preds, y)

    loss.backward()
    optimizer.step()

    if epoch % 5 == 0:
        print(f"Epochs {epoch} | Loss: {loss.item():.4f}")
```

3. NLP tie-in

```python
text = "the quick brown fox"
vocab = {w: i for i, w in enumerate(text.split())}

tokens = torch.tensor([vocab[w] for w in text.split()])
embed = nn.Embedding(len(vocab), 10)

embedded = embed(tokens).unsqueeze(0)  # batch-1
output = model(embedded)

print("RNN output: ", output)
```
