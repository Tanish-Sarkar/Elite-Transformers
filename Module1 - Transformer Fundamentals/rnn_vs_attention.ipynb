```python
import torch
import torch.nn as nn
import time

device = "cuda" if torch.cuda.is_available() else "cpu"

batch = 32
seq_len = 512
dim = 512

x = torch.randn(batch, seq_len, dim, device=device)

# -------------------------
# Dummy recurrence (RNN-like)
# -------------------------
W = torch.randn(dim, dim, device=device)
h = torch.zeros(batch, dim, device=device)

start = time.time()
for t in range(seq_len):
    h = torch.tanh(x[:, t, :] @ W + h)
torch.cuda.synchronize() if device == "cuda" else None
rnn_time = time.time() - start

# -------------------------
# Matrix-based attention
# -------------------------
Wq = torch.randn(dim, dim, device=device)
Wk = torch.randn(dim, dim, device=device)

start = time.time()
Q = x @ Wq
K = x @ Wk
scores = Q @ K.transpose(-2, -1)
torch.cuda.synchronize() if device == "cuda" else None
attn_time = time.time() - start

print(f"Dummy recurrence time: {rnn_time:.4f}s")
print(f"Matrix attention time: {attn_time:.4f}s")
```
